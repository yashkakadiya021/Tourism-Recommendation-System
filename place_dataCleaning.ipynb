```python
import pandas as pd
```

```python
df=pd.read_csv("place_to_visit.csv")
df
```

```python
import re

# Function to remove numbers from place names
def remove_numbers(place_name):
    return re.sub(r'^\d+\.\s*', '', place_name)  # Removes leading numbers followed by a dot and space

# Apply the function to the "places name" column
df["places name"] = df["places name"].apply(remove_numbers)

# Display the first few rows after cleaning
df.head()
```

```python
# Define category keywords
categories = {
    "Historical": ["fort", "palace", "museum", "heritage", "ruins", "historical", "monument", "memorial", "castle"],
    "Mountains": ["hill", "mountain", "peak", "trek", "valley", "range", "ghat", "hills"],
    "Beaches": ["beach", "island", "bay", "coast", "sea", "ocean", "shore"],
    "Wildlife/Nature": ["sanctuary", "national park", "wildlife", "forest", "waterfall", "lake", "dam", "garden", "zoo"],
    "Urban/Modern": ["mall", "market", "amusement", "park", "tower", "city", "museum", "stadium"],
    "Spiritual": ["temple", "church", "mosque", "gurudwara", "shrine", "ashram", "pilgrimage", "spiritual"]
}

# Function to categorize places based on names and descriptions
def classify_place(place_name, description):
    text = f"{place_name} {description}".lower()  # Combine both fields and convert to lowercase
    for category, keywords in categories.items():
        if any(keyword in text for keyword in keywords):
            return category
    return "Other"  # Default if no category matches

# Apply classification function
df["Category"] = df.apply(lambda row: classify_place(row["places name"], row["places descripation"]), axis=1)

# Display sample results
df.head()
```

```python
df.to_csv("clean_place.csv")
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('clean_place.csv')

# Function to append "from city center" after "km"
def append_city_center(distance):
    if isinstance(distance, str) and "km" in distance:
        return distance.replace("km", "km from city center")
    return distance

# Apply the function to the "distance" column
df['distance'] = df['distance'].apply(append_city_center)

# Save the updated CSV file
```
