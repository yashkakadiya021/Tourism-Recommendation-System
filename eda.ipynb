```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv("final_places.csv")
df.sample(10)
```

```python
h_df =pd.read_csv("final_hotel.csv")
h_df.sample(10)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(y = df["category"], order=df["category"].value_counts().index[:10])
for p in ax.patches:
    ax.annotate(f'{int(p.get_width())}', (p.get_width() + 1, p.get_y() + p.get_height()/2), 
                 ha='left', va='center', fontsize=12, color='black')
plt.title("Most Frequent Place Categories")
plt.xlabel("number of places")
plt.ylabel("Category")
plt.show()
```

```python
sns.histplot(df["Rating"].dropna(), bins=10, kde=True)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("numbers of places")
plt.show()
```

```python
top_places = df.groupby("Category")["Rating"].idxmax()
best_places = df.loc[top_places, ["Category", "places name", "Rating"]].sort_values(by="Rating", ascending=False)
print(best_places)
```

```python
plt.figure(figsize=(12, 6))
category_ratings = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)

sns.barplot(y=category_ratings.index, x=category_ratings.values, palette="viridis")
plt.title("Average Ratings by Place Category")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["distance"], y=df["Rating"], alpha=0.7)
plt.title("Distance vs Rating")
plt.xlabel("Distance (km)")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df["destination name"], order=df["destination name"].value_counts().index[:10])
plt.title("Top 10 Cities with the Most Tourist Attractions")
plt.xlabel("Number of Attractions")
plt.ylabel("City")
plt.show()
```

```python
import numpy as np  

plt.figure(figsize=(6, 4))
sns.heatmap(df[["Rating", "distance"]].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(h_df["rating"], bins=20, kde=True, color="blue")
plt.title("Distribution of Hotel Ratings")
plt.xlabel("rating")
plt.ylabel("number of hotel")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(y=h_df["destination name"], order=h_df["destination name"].value_counts().index[:10], palette="coolwarm")

# Add value labels
for p in ax.patches:
    ax.annotate(f'{int(p.get_width())}', (p.get_width() + 1, p.get_y() + p.get_height()/2), 
                 ha='left', va='center', fontsize=12, color='black')

plt.title("Top 10 Cities with the Most Hotels")
plt.xlabel("Number of Hotels")
plt.ylabel("City")
plt.show()
```

```python
h_df["discount price"] = h_df["actual price"] - h_df["discount price"]

plt.figure(figsize=(12, 6))
sns.boxplot(x="destination name", y="discount price", data=h_df)
plt.xticks(rotation=90)
plt.title("Hotel Discount Amounts in Different Cities")
plt.xlabel("City")
plt.ylabel("Discount Amount")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=h_df["discount price"], y=h_df["rating"], alpha=0.7)
plt.title("Hotel Price vs Ratings")
plt.xlabel("Discount Price")
plt.ylabel("Ratings")
plt.show()
```

```python
import numpy as np  

plt.figure(figsize=(6, 4))
sns.heatmap(h_df[["rating", "discount price", "actual price"]].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
